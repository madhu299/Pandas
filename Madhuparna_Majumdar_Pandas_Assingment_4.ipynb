```
#  Q1.- Make a pivot table of top 10 bowlers(in terms of wickets) with over(1-20) showing no. of wickets.  [Dificult-]
```

```
new=df2
new['wicket']=df2['dismissal_kind'].notnull()
new=new[new['wicket']==True]
best=new.groupby('bowler')['wicket'].sum().sort_values(ascending=False).head(10).index
new=new[new['bowler'].isin(best)]
new.pivot_table(index='bowler',columns='over',values='wicket',aggfunc='count')
```

```
# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]
```

Ex- 

V Kohli:

Innings:  

Total Runs: 

No. of MOM(Man of the match):

No. of boundries:

```
def profile(batsman_name):
  x=df2[df2['batsman']==batsman_name]
  print(batsman_name)
  print("Innings: ",x.drop_duplicates(subset=['match_id'],keep='first').shape[0])
  print("Total Runs: ",x['batsman_runs'].sum())
  print("No. of MOM(Man Of the match): ",df1[df1['player_of_match']==batsman_name].shape[0])
  print("No. of boundries: ",x[x['batsman_runs']==4].shape[0] + x[x['batsman_runs']==6].shape[0])
```

```
profile('V Kohli')
```

```
 # Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```
def top_batsman(venue_name):
  mega=df1.merge(df2,left_on='id',right_on='match_id')
  x=mega[mega['venue']==venue_name]
  print(x.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False).head(4))
```

```
top_batsman('Eden Gardens')
```
